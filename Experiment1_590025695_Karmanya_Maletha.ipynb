**Q6) Write a program to compute the length of the hypotenuse (c) of a right triangle
using Pythagoras theorem.**

```python
import math
p=int(input("Enter Side1:"))
b=int(input("Enter Side2:"))
sq = math.sqrt(p**2+b**2)
print(f"Length of Hypotenuse is = {sq:.2f}")
```

**Q8) Write a program to find area of triangle when length of sides are given.**

```python
import math
a=int(input("Enter Side1:"))
b=int(input("Enter Side2:"))
c=int(input("Enter Side3:"))
s = (a+b+c)/2
Area = math.sqrt(s*(s-a) * (s-b) * (s-c))
print(f"Area of triangle = {Area:.2f}")
```

**Q12) Write a program to print truth table for bitwise operators (&, | and ^ operators)**


```python

print(f"{'A':>3} {'B':>3} {'A&B':>5} {'A|B':>5} {'A^B':>5} {'A NAND B':>8} {'A NOR B':>8}\n")

for i in [0,1]:
    for j in [0,1]:
        print(f"{i:>3} {j:>3} {i & j:>5} {i | j:>5} {i^j :>5} {not(i & j):>8} {not(i | j ):>8}\n")
```
