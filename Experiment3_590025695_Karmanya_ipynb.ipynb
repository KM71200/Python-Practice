**Q1. Find a factorial of given number.**

```python
n = int(input("Enter a number to find factorial: "))
fact = 1
for i in range(1,n+1):
    fact = fact*i
print(fact)
```

**Q2) Find whether the given number is Armstrong number.**

```python
x = int(input("Enter a number"))
temp = x
s = 0
l = len(str(x))

while x!=0:
    r = x%10
    s += r**l
    x = x//10

if s == temp:
    print(f"{temp} is an armstrong number")

else:
    print(f"{temp} is not an armstrong nubmer")
```

**Q3) Print Fibonacci series up to given term.**

```python
n = int(input("Enter number:"))
a,b = 0,1
for _ in range(n):
    print(a," ")
    a,b = b, a+b
```

**Q4) Write a program to find if given number is prime number or not.**

```python
n = int(input("Enter number:"))
is_prime = True

if n<2:
    print(f"{n} is not a prime number")

else:
    for i in range(2,n):
        if n%i == 0:
            is_prime = False
            break

    if is_prime == True:
        print(f"{n} is Prime")

    else:
        print(f"{n} is not a prime number")
```

**Q5) Check whether given number is palindrome or not.**

```python
n = int(input("Enter a number:"))
temp = n
rev = 0
while n!=0:
  r = n%10
  rev = rev*10 + r
  n = n//10

if rev == temp:
  print(f"{temp} is a pallindrome number")

else:
  print(f"{temp} is not a pallindrome number")

```

**Q6) Write a program to print sum of digits**

```python
x = int(input("Enter a number:"))
sum = 0
while x!=0:
  r = x%10
  sum = sum+r
  x = x//10
print(f"Sum of digits = {sum}")
```

**Q7) Count and print all numbers divisible by 5 or 7 between 1 to 100.**

```python
count = 0
print("Numbers divisible by 5 or 7 are:")
for i in range(1,101):
  if i%5 == 0 or i%7 == 0:
    count += 1
    print(i,end = " ")
print(f"\nTotal numbers = {count}")
```
