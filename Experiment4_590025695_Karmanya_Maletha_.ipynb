**Q1.Write a program to count and display the number of capital letters in a given string.**

```python
s1 = input("Enter string:")
count = 0
for i in s1:
    if i.isupper():
        count = count + 1

print(f"Capiatl letters count = {count}")
```

**Q2) Count total number of vowels in a given string.**

```python
s = input("Enter string:")
count = 0
for i in s.lower():
    if i in {'a','e','i','o','u'}:
        count = count + 1
print(f"Total number of vowels = {count}")
```

**Q4) WAP to enter a string and a substring. You have to print the number of times that the substring occurs in the given string. String traversal will take place from left to right, not from right to left.**

```python
s = input("Enter String:")
sub = input("Enter Substring")
count = 0
start = 0
while True:
    pos = s.find(sub,start)
    if pos == -1:
        break
    count += 1
    start = pos + 1

print(f"Number of occurences of {sub}:{count}")
```

**5) Given a string containing both upper and lower case alphabets. Write a Python program to count the number of occurrences of each alphabet (case
insensitive) and display the same.**


```python
s = input("Enter a string:")
s1 = s.lower()
count = 0;
temp = ''
for word in s1:
    if word not in temp:
        count = s1.count(word)
        temp = temp + word
        print(f"{count}{word.upper()}")
```

**Q6) Program to count number of unique words in a given sentence using sets.**

```python
s = input("Enter String:").lower()
words = s.split()
s1 = set(words)
print(f"Number of unique words:{len(s1)}")
```

**Q7) Create 2 sets s1 and s2 of n fruits each by taking input from user and find:
a) Fruits which are in both sets s1 and s2
b) Fruits only in s1 but not in s2
c) Count of all fruits from s1 and s2**

```python
n = int(input("Enter number of fruits to enter:"))
s1 = set()
s2 = set()
print("Enter Elements for Set1:")
for i in range(n):
    element = input("Enter name of fruit:")
    s1.add(element)

print("Enter Elements for Set2:")
for j in range(n):
    element = input("Enter name of fruit:")
    s2.add(element)

print(f"Fruits in both:{s1.union(s2)}")
print(f"Fruit only in set 1:{s1.difference(s2)}")
print(f"Total cPount of fruits:{len(s1.union(s2))}")
```

**Q8. Take two sets and apply various set operations on them :
S1 = {Red ,yellow, orange , blue }
