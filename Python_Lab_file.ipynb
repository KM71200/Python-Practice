**Q5) Check whether the quadratic equation has real roots or imaginary roots.
Display the roots.**

```python
import math as m
b = float(input("Enter value of b:"))
a = float(input("Enter value of a:"))
c = float(input("Enter value of c:"))

D = b**2 - 4*a*c

if D > 0:
  r1 = (-b + m.sqrt(D/(2*a)))
  r2= (-b- m.sqrt(D/(2*a)))
  print("Roots are real and distinct")
  print(f"Roots are:{r1,r2}")

elif D==0:
  r1 = -b/(2*a)
  print("Roots are equal:")
  print(f"Roots are:{r1}")

else:
  real = -b/(2*a)
  comp = m.sqrt(-D/(2*a))
  print("Roots are complex")
  print(f"Roots are:{real},{comp:.2f}")
```
